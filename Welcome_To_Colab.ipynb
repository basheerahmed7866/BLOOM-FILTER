```
import numpy as np
import matplotlib.pyplot as plt

class NeuralNetwork:
    def __init__(self, learning_rate):
        self.weights = np.array([np.random.randn(), np.random.randn()])
        self.bias = np.random.randn()
        self.learning_rate = learning_rate

    def _sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def _sigmoid_deriv(self, x):
        return self._sigmoid(x) * (1 - self._sigmoid(x))

    def predict(self, input_vector):
        layer_1 = np.dot(input_vector, self.weights) + self.bias
        layer_2 = self._sigmoid(layer_1)
        prediction = layer_2
        return prediction

    def _compute_gradients(self, input_vector, target):
        layer_1 = np.dot(input_vector, self.weights) + self.bias
        layer_2 = self._sigmoid(layer_1)
        prediction = layer_2
        derror_dprediction = 2 * (prediction - target)
        dprediction_dlayer1 = self._sigmoid_deriv(layer_1)
        dlayer1_dbias = 1
        dlayer1_dweights = input_vector
        derror_dbias = derror_dprediction * dprediction_dlayer1 * dlayer1_dbias
        derror_dweights = derror_dprediction * dprediction_dlayer1 * dlayer1_dweights
        return derror_dbias, derror_dweights

    def _update_parameters(self, derror_dbias, derror_dweights):
        self.bias = self.bias - (derror_dbias * self.learning_rate)
        self.weights = self.weights - (derror_dweights * self.learning_rate)

    def train(self, input_vectors, targets, iterations):
        cumulative_errors = []
        for current_iteration in range(iterations):
            # Pick a data instance at random
            random_data_index = np.random.randint(len(input_vectors))
            input_vector = input_vectors[random_data_index]
            target = targets[random_data_index]
            # Compute the gradients and update the weights
            derror_dbias, derror_dweights = self._compute_gradients(input_vector, target)
            self._update_parameters(derror_dbias, derror_dweights)
            # Measure the cumulative error for all the instances
            if current_iteration % 100 == 0:
                cumulative_error = 0
                # Loop through all the instances to measure the error
                for data_instance_index in range(len(input_vectors)):
                    data_point = input_vectors[data_instance_index]
                    target = targets[data_instance_index]
                    prediction = self.predict(data_point)
                    error = np.square(prediction - target)
                    cumulative_error += error
                cumulative_errors.append(cumulative_error)
        return cumulative_errors

    def plot_neural_network(self):
        fig, ax = plt.subplots(figsize=(10, 6))
        # Nodes positions
        input_layer = np.array([[1, 3], [1, 2]])
        hidden_layer = np.array([[3, 2.5]])
        output_layer = np.array([[5, 2.5]])
        # Plot nodes
        ax.scatter(input_layer[:, 0], input_layer[:, 1], s=500, label='Input Layer')
        ax.scatter(hidden_layer[:, 0], hidden_layer[:, 1], s=500, label='Hidden Layer')
        ax.scatter(output_layer[:, 0], output_layer[:, 1], s=500, label='Output Layer')
        # Plot edges and weights
        for i in range(len(input_layer)):
            ax.plot([input_layer[i, 0], hidden_layer[0, 0]], [input_layer[i, 1], hidden_layer[0, 1]], 'k-')
            ax.text(2, (input_layer[i, 1] + hidden_layer[0, 1]) / 2, f'w{i+1}={self.weights[i]:.2f}', fontsize=12)
        # Bias to output layer
        ax.plot([hidden_layer[0, 0], output_layer[0, 0]], [hidden_layer[0, 1], output_layer[0, 1]], 'k-')
        ax.text(4, (hidden_layer[0, 1] + output_layer[0, 1]) / 2, f'b={self.bias:.2f}', fontsize=12)
        # Annotate nodes
        for i in range(len(input_layer)):
            ax.text(input_layer[i, 0] - 0.1, input_layer[i, 1], f'Input {i+1}', fontsize=12, ha='right')
        ax.text(hidden_layer[0, 0] + 0.1, hidden_layer[0, 1], 'Neuron', fontsize=12, ha='left')
        ax.text(output_layer[0, 0] + 0.1, output_layer[0, 1], 'Output', fontsize=12, ha='left')
        # Display the plot
        ax.set_xlim(0, 6)
        ax.set_ylim(1, 3.5)
        ax.axis('off')
        plt.legend(loc='best')
        plt.title("Neural Network Diagram")
        plt.show()

# Initialize input data and targets
input_vectors = np.array([
    [3, 1.5],
    [2, 1],
    [4, 1.5],
    [3, 4],
    [3.5, 0.5],
    [2, 0.5],
    [5.5, 1],
    [1, 1],
])
targets = np.array([0, 1, 0, 1, 0, 1, 1, 0])

# Train the neural network
learning_rate = 0.1
neural_network = NeuralNetwork(learning_rate)
training_error = neural_network.train(input_vectors, targets, 10000)

# Plot the training error
plt.plot(training_error)
plt.xlabel("Iterations (x100)")
plt.ylabel("Cumulative Error")
plt.title("Cumulative Error vs. Iterations")
plt.savefig("cumulative_error.png")
plt.show()

# Plot the neural network diagram with weights and bias
neural_network.plot_neural_network()
```

```
import numpy as np
import matplotlib.pyplot as plt

class NeuralNetwork:
    def __init__(self, learning_rate):
        self.weights = np.array([np.random.randn(), np.random.randn()])
        self.bias = np.random.randn()
        self.learning_rate = learning_rate

    def _sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def _sigmoid_deriv(self, x):
        return self._sigmoid(x) * (1 - self._sigmoid(x))

    def predict(self, input_vector):
        layer_1 = np.dot(input_vector, self.weights) + self.bias
        layer_2 = self._sigmoid(layer_1)
        prediction = layer_2
        return prediction

    def _compute_gradients(self, input_vector, target):
        layer_1 = np.dot(input_vector, self.weights) + self.bias
        layer_2 = self._sigmoid(layer_1)
        prediction = layer_2
        derror_dprediction = 2 * (prediction - target)
        dprediction_dlayer1 = self._sigmoid_deriv(layer_1)
        dlayer1_dbias = 1
        dlayer1_dweights = input_vector
        derror_dbias = derror_dprediction * dprediction_dlayer1 * dlayer1_dbias
        derror_dweights = derror_dprediction * dprediction_dlayer1 * dlayer1_dweights
        return derror_dbias, derror_dweights

    def _update_parameters(self, derror_dbias, derror_dweights):
        self.bias = self.bias - (derror_dbias * self.learning_rate)
        self.weights = self.weights - (derror_dweights * self.learning_rate)

    def train(self, input_vectors, targets, iterations):
        cumulative_errors = []
        for current_iteration in range(iterations):
            # Pick a data instance at random
            random_data_index = np.random.randint(len(input_vectors))
            input_vector = input_vectors[random_data_index]
            target = targets[random_data_index]
            # Compute the gradients and update the weights
            derror_dbias, derror_dweights = self._compute_gradients(input_vector, target)
            self._update_parameters(derror_dbias, derror_dweights)
            # Measure the cumulative error for all the instances
            if current_iteration % 100 == 0:
                cumulative_error = 0
                # Loop through all the instances to measure the error
                for data_instance_index in range(len(input_vectors)):
                    data_point = input_vectors[data_instance_index]
                    target = targets[data_instance_index]
                    prediction = self.predict(data_point)
                    error = np.square(prediction - target)
                    cumulative_error += error
                cumulative_errors.append(cumulative_error)
        return cumulative_errors

    def get_predictions(self, input_vectors):
        predictions = []
        for input_vector in input_vectors:
            predictions.append(self.predict(input_vector))
        return np.array(predictions)

    def plot_neural_network(self):
        fig, ax = plt.subplots(figsize=(10, 6))
        # Nodes positions
        input_layer = np.array([[1, 3], [1, 2]])
        hidden_layer = np.array([[3, 2.5]])
        output_layer = np.array([[5, 2.5]])
        # Plot nodes
        ax.scatter(input_layer[:, 0], input_layer[:, 1], s=500, label='Input Layer')
        ax.scatter(hidden_layer[:, 0], hidden_layer[:, 1], s=500, label='Hidden Layer')
        ax.scatter(output_layer[:, 0], output_layer[:, 1], s=500, label='Output Layer')
        # Plot edges and weights
        for i in range(len(input_layer)):
            ax.plot([input_layer[i, 0], hidden_layer[0, 0]], [input_layer[i, 1], hidden_layer[0, 1]], 'k-')
            ax.text(2, (input_layer[i, 1] + hidden_layer[0, 1]) / 2, f'w{i+1}={self.weights[i]:.2f}', fontsize=12)
        # Bias to output layer
        ax.plot([hidden_layer[0, 0], output_layer[0, 0]], [hidden_layer[0, 1], output_layer[0, 1]], 'k-')
        ax.text(4, (hidden_layer[0, 1] + output_layer[0, 1]) / 2, f'b={self.bias:.2f}', fontsize=12)
        # Annotate nodes
        for i in range(len(input_layer)):
            ax.text(input_layer[i, 0] - 0.1, input_layer[i, 1], f'Input {i+1}', fontsize=12, ha='right')
        ax.text(hidden_layer[0, 0] + 0.1, hidden_layer[0, 1], 'Neuron', fontsize=12, ha='left')
        ax.text(output_layer[0, 0] + 0.1, output_layer[0, 1], 'Output', fontsize=12, ha='left')
        # Display the plot
        ax.set_xlim(0, 6)
        ax.set_ylim(1, 3.5)
        ax.axis('off')
        plt.legend(loc='best')
        plt.title("Neural Network Diagram")
        plt.show()

# Initialize input data and targets
input_vectors = np.array([
    [3, 1.5],
    [2, 1],
    [4, 1.5],
    [3, 4],
    [3.5, 0.5],
    [2, 0.5],
    [5.5, 1],
    [1, 1],
])
targets = np.array([0, 1, 0, 1, 0, 1, 1, 0])

# Train the neural network
learning_rate = 0.1
neural_network = NeuralNetwork(learning_rate)
training_error = neural_network.train(input_vectors, targets, 10000)

# Plot the training error
plt.plot(training_error)
plt.xlabel("Iterations (x100)")
plt.ylabel("Cumulative Error")
plt.title("Cumulative Error vs. Iterations")
plt.savefig("cumulative_error.png")
plt.show()

# Get and print the predicted output values after training
predicted_outputs = neural_network.get_predictions(input_vectors)
print("Predicted outputs after training:")
print(predicted_outputs)

# Plot the neural network diagram with weights and bias
neural_network.plot_neural_network()
```
